```python
import os
import json
os.chdir("output/fhir")
```

```python
fileList = os.listdir()
```

```python
entryList = []
for f in fileList:
    fOpen = open(f)
    print(f)
    try:
        fJson = json.load(fOpen)
    except:
        continue
    entryList.append(fJson['entry'])
    fOpen.close()
```

```python
bPList = [(entry['resource']['component'][0]['valueQuantity']['value'],entry['resource']['component'][1]['valueQuantity']['value']) for entry in entryList[0] if entry['resource']['resourceType'] == "Observation" and entry['resource']['code']['coding'][0]['display'] == "Blood Pressure"]
    
```

```python
totalBPListX = []
totalBPListY = []
totalBPList = []
for patient in entryList:
    bPList = [(entry['resource']['component'][0]['valueQuantity']['value'],entry['resource']['component'][1]['valueQuantity']['value']) for entry in patient if entry['resource']['resourceType'] == "Observation" and entry['resource']['code']['coding'][0]['display'] == "Blood Pressure"]
    sum1, sum2 = (0, 0)
    for recording in bPList:
        cur1, cur2 = recording
        sum1 += cur1
        sum2 += cur2
    try:
        sum1 /= len(bPList)
        sum2 /= len(bPList)
        totalBPListX.append(sum1)
        totalBPListY.append(sum2)
        totalBPList.append((sum1, sum2, patient))
    except:
        continue
```

```python
highBPList = []
for x, y, patient in totalBPList:
    if y > 140:
        patCondList = [entry['resource']['code']['text'] for entry in patient if entry['resource']['resourceType'] == "Condition"]
        highBPList.append(patCondList)
```

```python
len(highBPList)
```

```python
diseaseDict = {}
for patient in highBPList:
    currentFound = []
    for cond in patient:
        if cond not in currentFound:
            try:
                diseaseDict[cond] += 1
            except:
                diseaseDict[cond] = 1
    print(cond)
```
